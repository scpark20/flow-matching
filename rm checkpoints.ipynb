```python
import os

dir = '/data/openai_log/64x64_fm_warm5'
files = [os.path.join(dir, f) for f in os.listdir(dir)]
files = sorted(files)
files
```

```python
#   
files_to_delete = []
files_to_save = []
keywords = ['model', 'ema_0.9999_', 'opt']
#    
for file in files:
    if '.pt' not in file:
        continue
    
    step_number = None
    for keyword in keywords:
        if keyword in file:
            step_number = int(file.split(keyword)[-1].split('.')[0])
    if step_number is None:
        continue
    
    if step_number % 10000 != 0:
        files_to_delete.append(file)
    else:
        files_to_save.append(file)
        

#     (  )
for file in files_to_delete:
    print("Deleting:", file)
    os.remove(file)  #      .
```
